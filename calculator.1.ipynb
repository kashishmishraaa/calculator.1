```python
import tkinter as tk
from functools import partial

def calculate(event=None):
    try:
        result = str(eval(display.get()))
        display.delete(0, tk.END)
        display.insert(tk.END, result)
    except Exception as e:
        display.delete(0, tk.END)
        display.insert(tk.END, "Error")

def press(key):
    display.insert(tk.END, key)

def clear():
    display.delete(0, tk.END)

root = tk.Tk()
root.title("Calculator")

# Get the screen width and height
screen_width = root.winfo_screenwidth()
screen_height = root.winfo_screenheight()

# Set window size
window_width = 400
window_height = 600

# Calculate x and y coordinates for the Tk root window
x = (screen_width / 2) - (window_width / 2)
y = (screen_height / 2) - (window_height / 2)

# Set the dimensions of the screen and where it is placed
root.geometry(f"{window_width}x{window_height}+{int(x)}+{int(y)}")

# Display
display = tk.Entry(root, font=("Arial", 24), bd=10, insertwidth=2, width=14, borderwidth=4)
display.grid(row=0, columnspan=4, padx=10, pady=10)

button_parameters_black = {
    'fg': 'white',
    'bg': 'black',
    'font': ('Arial', 18),
    'bd': 5,
    'width': 5,
    'height': 2
}

button_parameters_white = {
    'fg': 'black',
    'bg': 'white',
    'font': ('Arial', 18),
    'bd': 5,
    'width': 5,
    'height': 2
}

button_parameters_gray = {
    'fg': 'black',
    'bg': 'gray',
    'font': ('Arial', 18),
    'bd': 5,
    'width': 5,
    'height': 2
}

number_buttons = ['7', '8', '9', '4', '5', '6', '1', '2', '3', '0']
symbol_buttons = ['/', '*', '-', '+', '.']
buttons = [
    ('7', 1, 0), ('8', 1, 1), ('9', 1, 2), ('/', 1, 3), 
    ('4', 2, 0), ('5', 2, 1), ('6', 2, 2), ('*', 2, 3),
    ('1', 3, 0), ('2', 3, 1), ('3', 3, 2), ('-', 3, 3),
    ('0', 4, 0), ('.', 4, 1), ('+', 4, 3), ('=', 4, 2)
]

for (text, row, col) in buttons:
    if text == '=':
        button = tk.Button(root, text=text, command=calculate, **button_parameters_gray)
        root.bind('<Return>', calculate)
    elif text in number_buttons:
        button = tk.Button(root, text=text, command=partial(press, text), **button_parameters_black)
    else:
        button = tk.Button(root, text=text, command=partial(press, text), **button_parameters_white)
    button.grid(row=row, column=col, padx=10, pady=10, sticky='nsew')

clear_button = tk.Button(root, text="C", command=clear, **button_parameters_white)
clear_button.grid(row=4, column=3, padx=10, pady=10, sticky='nsew')

root.mainloop()
```
